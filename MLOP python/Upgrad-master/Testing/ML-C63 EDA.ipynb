```python
def create_item_type(data_frame):
  df['Item_Type'] = df['Item_Identifier'].str[:2]
  df['Item_Type'] = df['Item_Type'].map({
                                            'FD' : 'Food',
                                            'NC' : 'Non_Consumables',
                                            'DR' : 'Drink'
                            })
  return data_frame
```

```python
df = create_item_type(df)
df.head()
```

```python

def standardize_item_fat_content(data_frame):
  df['Item_Fat_Content'] = df['Item_Fat_Content'].replace({
                                            'Low Fat' : 'Low_Fat',
                                            'LF' : 'Low_Fat',
                                            'reg' : 'Regular',
                                            'low fat' : 'Low_Fat'
                                   })
  return data_frame
```

```python
df = standardize_item_fat_content(df)
df.head()
```
